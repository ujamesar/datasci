```python

# Q1  Q1. Explain with an example each when to use a for loop and a while loop.

# The major difference between for loop and the while loop is that for loop is used when the number of iterations is known,
# whereas execution is done in the while loop until the statement in the program is proved wrong.


# While loop
n = 0
while n < 10: # while n is less than 10,  Here it checks the condition from above value of n then proceeds further
    print(n) # print out the value of n 
    n += 1   # Changes the value to n+1 and loop again go to check at above condition
             # it is running till n<10 means it runs till statement proves wrong
    
print("     ")   
# for loop
for i in range(10): # here it iterates till range as range is 
    print(i)
```

```python
# Q2 Write a python program to print the sum and product of the first 10 natural numbers using for
#and while loop.

#with while loop

num=10
product=1
sum=0
while(num > 0):
    sum += num
    product=product * num
    num -= 1
print("The sum is", sum)
print("product is " , product)
```

```python
#Q3. Create a python program to compute the electricity bill for a household.
#The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
#unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
#be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
#You are required to take the units of electricity consumed in a month from the user as input.
#Your program must pass this test case: when the unit of electricity consumed by the user in a month is
#310, the total electricity bill should be 2250.
unit=int(input("Input monthly used unit to know the charge"))
while (unit>0):
    if unit<=100:
        print(unit*4.5)
        break
    elif unit >100 and unit <=200:
        print((100*4.5 )+ (unit-100)*6)
        break
    
    elif unit >200 and unit <=300:
        print((100*4.5)+(100*6)+(unit-200)*10)
        break  
    elif unit >300:
        print("Total Electricity bill is ",(100*4.5)+(100*6)+(100*10)+(unit-300)*20)
        break 
```

```python
#Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
#number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
#that list.


# with for loop
l=[]
l2=[]
for i in range(1,101):
    l.append(i)
for i in l:
    print(f"Cube of {i} is {i*i*i} ")
    if (i*i*i)%4==0 or (i*i*i)%5==0:
        l2.append(i)
print("New list =",l2) 
```
